```python
from owslib.wms import WebMapService
wms=WebMapService('https://maps.planet.fu-berlin.de/eqc-bin/wms', version='1.1.1')
#wms.identification.type
#wms.identification.version
#wms.identification.title
#wms.contents
#for layer in list(wms.contents):
#    print(layer, wms[layer])
#print(wms['MOLA-color-hs'].boundingBox)
print(wms['MOLA-color-hs'].crsOptions)
#print(wms.getOperationByName('GetMap').formatOptions)
```

```python
from IPython.display import Image
layer='HMChsv'
img = wms.getmap(layers=[layer],
                 #srs='IAU2000:49910',
                 srs=wms[layer].crsOptions[-1],
                 bbox=wms[layer].boundingBox[0:-1],
                 size=(1200,800),
                 format='image/png',
                 transparent=True
                )
print(img.info())

out = open('hrsc_hsv.png', 'wb')
out.write(img.read())
Image(filename='hrsc_hsv.png')
```
